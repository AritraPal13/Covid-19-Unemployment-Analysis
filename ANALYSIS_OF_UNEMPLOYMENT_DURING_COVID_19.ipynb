##**ANALYSIS OF UNEMPLOYMENT DURING COVID 19**

##**INTRODUCTION**

**What is unemployment?** 

**The term unemployment refers to a state in which an individual actively seeks employment but is unsuccessful.**


**To analyse the unemployment condition during Covid 19,I will be using a dataset that contains the data of unemployment during C-19 of India and further we are going analyse the states of India to get the unemployment rate.**

##**AIM:**
**The main aim of the analysis is to know the unemployment rate and status during COVID-19 period of India.**

##**FEATURES:**

###**States**
**It is defined as the name of the states of the Indian nation.**

###**Date**
**It is the date on which the information was extracted.**

**The records are taken for each months of the year for every state.**

###**Estimated Unemployment Rate**
**It is the unemployment rate of each state in that segment of month of that year.**

###**Estimated Employed**
**It is the total number of peoples employed in that segment of month of that year for that respective state.** 

###**Estimated Labour Participation Rate**
**It is the rate of  total participation of the labour community of that state for that period.**

###**Region**
**It is the  8 Quadrants of the country.**

###**Longitude**
**A measure of how far west region is.**

###**Latitude**
**A meausre of how far north the region is.**

##**Lets Start The Analysis:**

**The dataset I’m using here contains data on India’s unemployment rate during Covid-19. So let’s start the task of Unemployment analysis by importing the necessary Python libraries and the dataset:**

```python
import numpy as np  # importing required libraries
import pandas as pd #numpy #pandas #seaborn #matplotlib
import seaborn as sns
import matplotlib.pyplot as plt  
import plotly.express as pex
coviddata = pd.read_csv("https://raw.githubusercontent.com/AritraPal13/Covid-19-Unemployment-Analysis/main/CovidUnemployment.csv")
coviddata
```

```python
coviddata.describe(include="all")  #describe(include='all') shows all the statiscal calculation in the data set.
```

```python
coviddata.dtypes     #.dtypes show all the data type of the features in the data set.
```

```python
h1=coviddata.isnull().sum()           #Using isnull() to select all rows with NaN under a single DataFrame column:
h1                                    #.sum() is used to get the null values of the respective columns
```

**From the above analysis we can say there are no missing values in the dataset.**

**But as we go further analysis there may be some problems with column name so we are going to rename them for future fault prevention.**

```python
coviddata.columns= ["States","Date",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
```

###**Analysis 1:**

**Lets show with the help of visualization Unemployment Rate according to Region of India**

```python
plt.figure(figsize=(15,10))
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=coviddata).set(title="Unemployment rate according to Region")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.displot(coviddata, x="Estimated Unemployment Rate",hue="Region", kind="kde",height=8).set(title="Unemployment Rate according to Region")
plt.grid(True)
plt.show
```

###**Analysis 2:**

**Now let’s see the Estimated No of Employee according to different Regions of India:**

```python
plt.figure(figsize=(15,10))
sns.histplot(x="Estimated Employed",hue="Region",data=coviddata).set(title="Estimated No of Employee according to Region")
plt.grid(True)
plt.show()
```

```python
sns.displot(coviddata, x="Estimated Employed",hue="Region", kind="kde",height=8).set(title="Estimated No of Employee according to Regions")
plt.grid(True)
plt.show
```

###**Analysis 3:**

**Lets show with the help of visualization Unemployment Rate according to Indian State**

```python
fig = pex.sunburst(coviddata[["States", "Estimated Unemployment Rate"]], path=[ "States"], values="Estimated Unemployment Rate",
title="Unemployment Rate according to Indian States",height=800)
fig.show()
```

**From above analysis we can note that Haryana State has the highest unemployment rate i.e 274.77.**

**While Meghalaya State has the lowest unemployment rate in the country i.e 38.66**


###**Analysis 4:**

**Lets show with the help of visualization Estimated No of Employee according to Indian State**

```python
fig = pex.sunburst(coviddata[["States", "Estimated Employed"]], path=[ "States"], values="Estimated Employed",
title="Estimated No of Employee according to Indian States",height=800)
fig.show()
```

**From the above analysis we can note that Uttar Pradesh State recruited highest number of employees which is almost 555,247,990**
**While Sikkim state recruited lowest number of employees which is almost 1,876,186**

###**Analysis 5:**
**Now lets see the unemployment rate with respect to regions and states of India.**


```python
fig = pex.sunburst(coviddata[["States","Region", "Estimated Unemployment Rate"]], path=[ "Region","States"], values="Estimated Unemployment Rate",
title="Unemployment Rate with respect to Regions & States of India",height=800)
fig.show()
```

**From the above figure we can see the distribution of unemployment rate with respect to regions and states in that reggion.**

**We can see that north region states has the highest unemployment rate while West region has the lowest unemployment rate.**
